```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

#Inserir os tickers conforme o Yahoo Finance
TickerA= 'ITSA3.SA'
TickerB= 'PETR4.SA'
TickerC = 'ABEV3.SA'
TickerD = 'VIIA3.SA'
TickerE = 'VALE3.SA'
TickerF = 'BPFF11.SA'
TickerG = 'COGN3.SA'
TickerH = 'BBDC4.SA'
TickerI = 'SMAL11.SA'
prices=pd.DataFrame() 
tickers = [TickerA, TickerB, TickerC, TickerD,TickerE,TickerF,TickerG,TickerH, TickerI] #Inserir tickers
for t in tickers:
  data = yf.Ticker(t)
  prices[t]= data.history(period="512d")['Close'].dropna()
prices

```

```python
import matplotlib.pyplot as plt #Pacote de gráficos

(prices/prices.iloc[0]*100).plot(figsize=(30,10))
plt.ylabel('Preços Normalizados')
plt.xlabel('Período de 504d')
plt.show()
```

```python
log_returns = np.log(prices/prices.shift(1))
log_returns =log_returns
log_returns.to_excel('retornos.xls')
log_returns
prices.to_excel('prices.xls')
log_returns.dropna()
```

## GARCH(1,1)

```python
from arch import arch_model
ret = log_returns['PETR4.SA'].dropna()
model = arch_model(ret, vol='Garch', p=1, o=0, q=1, dist='Normal')
results = model.fit()
print(results.summary())
```

```python
# one-step out-of sample forecast
garch_forecast = results.forecast(horizon=1)
predicted_et = garch_forecast.mean['h.1'].iloc[-1]
predicted_et
```

### Simulação de Pesos para Carteira de Ativos

```python

np.random.seed(1)
nticks = len(tickers)
nsample=10000
pfolio_returns = []
pfolio_volatilities = []
all_weights = np.zeros((nsample, nticks))
exp_return = np.zeros(nsample)
exp_std = np.zeros(nsample)
exp_var = np.zeros(nsample)
Sharpe = np.zeros((nsample,1))
for i in range(nsample):
    weights = np.random.random(nticks)
    weights = weights/np.sum(weights)
    
    all_weights[i,:] = weights #salva os pesos
    
    exp_return[i]= np.sum(weights*log_returns.mean())*250 
    exp_var[i] = np.dot(weights.T,np.dot(log_returns.cov()*250, weights))
    exp_std[i] = np.sqrt(exp_var[i]) #Risco
    
    Sharpe[i,:]=exp_return[i]/exp_std[i] #Retorno/Risco
    
    pfolio_returns.append(exp_return[i])
    pfolio_volatilities.append(exp_std[i])
    
pfolio_returns=np.array(pfolio_returns)
pfolio_volatilities=np.array(pfolio_volatilities)
```

# Pesos da Carteira de Máximo Sharpe

```python

col = tickers[:]
col.append('sharpe')
df = pd.DataFrame(np.concatenate((all_weights,Sharpe), axis = 1),columns =col)
df.sort_values(by = ['sharpe'], ascending = False, inplace = True)
coluna = ['data','pesos']
df_sharpe= pd.DataFrame(df.iloc[0]).iloc[0:-1]
df_sharpe.columns = ['pesos (%)']
df_sharpe*100

```

```python
## Salientar o pedaço de maior proporção de alocação de ativos
optimal_weights = df.iloc[0,0:-1]
quant_tickers = len(tickers)

#Construir o explode do pie plot
explode = []
item_max = np.where(df.iloc[0,0:-1] == df.iloc[0,0:-1].max())

for x in range(quant_tickers):
  if x == item_max[0]:
    explode.append(0.2)
  else:
    explode.append(0)
explode = tuple(explode)
fig1, ax1 = plt.subplots(figsize = (15,15))
ax1.pie(optimal_weights, explode = explode, labels = tickers, autopct = '%1.1f%%')
ax1.axis('equal')
plt.show()
```

# Pesos da Carteira da Risco Mínimo


```python

min = exp_std.min()
item = np.where(exp_std == min)
optimal_weights = all_weights[item]*100
tickers_sem_ibov = np.transpose(tickers)

## Salientar o pedaço de maior proporção de alocação de ativos
max = optimal_weights.max()
item_max = np.where(optimal_weights == max)
quant_tickers = len(tickers_sem_ibov)

#Construir o explode do pie plot
explode = []
for x in range(quant_tickers):
  if x == item_max[1]:
    explode.append(0.1)
  else:
    explode.append(0)
explode = tuple(explode)


fig1, ax1 = plt.subplots(figsize = (15,15))
ax1.pie(optimal_weights, explode = explode, labels = tickers_sem_ibov, autopct = '%1.1f%%')
ax1.axis('equal')
plt.show()
```
