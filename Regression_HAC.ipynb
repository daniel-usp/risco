```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

fonte = 'https://raw.githubusercontent.com/daniel-usp/LIT/master/Exemplo_Reg.csv'
df = pd.read_csv(fonte, sep = ';')

reg = smf.ols('Q ~ P',data=df).fit(cov_type='HAC',cov_kwds={'maxlags':1})
print(reg.summary())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme(color_codes=True)
cm = 1/2.54  # centimeters in inches
plt.subplots(figsize=(30*cm, 15*cm))
ax = sns.regplot(x="P", y="Q", data=df)
```
